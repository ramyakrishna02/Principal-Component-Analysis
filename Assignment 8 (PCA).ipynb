### Importing required libraries 

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the dataset
WINE = pd.read_csv("wine.csv")
WINE.head()
```

```python
# to get the information of the datafile
WINE.info()
```

```python
# to get the values count of the column 'Type'
WINE['Type'].value_counts()
```

```python
# data without the 'Type' column
wine = WINE.iloc[:,1:]
wine
```

```python
# summarizing the datafile
wine.describe()
```

```python
# to check the null values of the data
wine.isnull().sum()
```

```python
print("Dimensions -",wine.shape)
print("Column names -",wine.columns)
```

# PCA

Principle Component Analysis 

```python
# Standardizing the numerical data
std = scale(wine)
std
```

```python
# Application of the PCA function on the data
pca = PCA(n_components=13)
pca_values = pca.fit_transform(std)
pca_values
```

```python
# Loadings or Weights 
pca.components_
```

```python
# The amount of variance that each PCA explains
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.figure(figsize=(5,3))
plt.plot(var1,color="blue")
```

```python
# to print the first 3 principle component scores
pca_values[:,0:3]
```

```python
# Creating final dataframe by combining the column 'Type' and first 3 principle component scores 
finaldf = pd.concat([WINE[["Type"]],pd.DataFrame(pca_values[:,:3],columns=['pc1','pc2','pc3'])],axis=1)
finaldf
```

```python
# plotting the final dataframe
plt.style.use('classic')
plt.figure(figsize=(13,8))
sns.scatterplot(data=finaldf)
plt.show()
```

# Hierarchical Clustering 

```python
# importing required libraries 
from sklearn.preprocessing import MinMaxScaler
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
from sklearn.cluster import AgglomerativeClustering # Hierarchical Clustering
```

```python
# Creating Dendrogram Structure 

z = linkage(std, method="ward",metric='euclidean')
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_rotation=1.0)
plt.show()
```

```python
# Creating Hierarchical cluster
h_clus = AgglomerativeClustering(n_clusters=3,linkage='ward',
                                 affinity='euclidean').fit(std)
clust_labels = pd.Series(h_clus.labels_)
clust_labels

# creating new column and assigning it to new column
clus_df = WINE.copy()
clus_df['Cluster'] = clust_labels
clus_df.tail(15)
```

```python
# using groupby func for the summarization of data by clusters
clus_df.iloc[:,1:].groupby(clus_df.Cluster).mean()
```

# K-Means Clustering

```python
# Importing KMeans function from sklearn library
from sklearn.cluster import KMeans
```

```python
# Elbow Curve for k-value
# plotting using ELBOW method to select best k value

fig = plt.figure(figsize=(5,3))
# WCSS - within cluster sum of square
WCSS = []
for i in range(1,8):
    clf = KMeans(n_clusters=i)
    clf.fit(std)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1,8),WCSS)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Creating K-Means cluster using k=3

# Prediction of the clusters to which the observations belong
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(std)
y_kmeans
```

```python
# to get the centers of the clusters
clf.cluster_centers_
```

```python
# to get the within cluster sum of square variance
clf.inertia_
```

```python
# converting numpy array into pandas series object
a = pd.Series(y_kmeans)
clus_df1 = WINE.copy()
clus_df1['Cluster'] = a
clus_df1
```

```python
# using groupby func for the summarization of data by clusters
clus_df1.iloc[:,1:].groupby(clus_df1.Cluster).mean()
```
